#Calculations

```python
# TASK 1

import numpy as np

R = 15
phi = np.pi/6
v_Mtr = 5*np.pi/2
V_Mrel = 27

v_My = v_Mtr*np.sin(np.pi/2-phi)
v_Mx = v_Mtr*np.cos(np.pi/2-phi) + V_Mrel
V_Mtot = (v_My**2 + v_Mx**2)**0.5


a_Mrel = 24
epsilon = np.pi/12
omega = np.pi/6

a_t = epsilon*R
a_ntr = omega**2*R

a_My = a_t*np.sin(np.pi/2 - phi) - a_ntr*np.sin(phi)
a_Mx = a_Mrel + a_t*np.cos(np.pi/2 - phi) + a_ntr*np.cos(phi)

a_Mtot = (a_My**2 + a_Mx**2)**0.5

print("Total velocity V_M", V_Mtot)
print("Total acceleratoins a_Mtot", a_Mtot)
```

#Simulation

```python
from sympy import *
from sympy import Symbol, Ellipse, Point, Line, Circle, Matrix, diff
from sympy.vector import CoordSys3D
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import numpy as np
```

```python
R = 15
O1A = R
O2B = R
SA = SB = 15/2
a = 15
SQ = QO = TR = ST = a

t_vec = np.linspace(0, np.pi/6, 130)
phi_vec = [2*t for t in t_vec]
sr_vec = [2*t**3+3*t for t in phi_vec]


O2 = Point(7.5, 0)
O1 = Point(0, 7.5)

A_vec = [O1A * Point([np.cos(t), np.sin(t)]) + O1 for t in phi_vec]
B_vec = [O2B * Point([np.cos(t), np.sin(t)]) + O2 for t in phi_vec]

S_vec = [Circle(A, SA).intersection(Circle(B, SB))[0] for A, B in zip(A_vec, B_vec)]

# Lay off the vector SQ in the direction of vector SA
SA_vec = [A - S for A, S in zip(A_vec, S_vec)]
Q_vec = [S + 2*SA for S, SA in zip(S_vec, SA_vec)]

# Lay off the vector ST in the direction of vector SB
SB_vec = [B - S for B, S in zip(B_vec, S_vec)]
T_vec = [S + 2*SB for S, SB in zip(S_vec, SB_vec)]

O_vec = [Circle(Q, QO).intersection(Circle(T, TR))[1] for Q, T in zip(Q_vec, T_vec)]

QO_vectors = [Q - O for Q, O in zip(Q_vec, O_vec)]
M_vec = [QO_vector*(sr/QO) + O  for QO_vector, O, sr in zip(QO_vectors, O_vec, sr_vec)]


fixed_points_with_names = [(O1, "O1"), (O2, "O2")]
```

```python
# SIMULATION

points_names = ["A", "B", "S", "T", "Q", "O", "M"]

positions= zip(A_vec, B_vec, S_vec, T_vec, Q_vec, O_vec, M_vec)
# velocities = zip(vel_A, vel_B, vel_C, vel_D, vel_E, vel_F)
# accelerations = zip(acc_A, acc_B, acc_C, acc_D, acc_E, acc_F)
states = positions

O2 = Point(7.5, 0)
O1 = Point(0, 7.5)


fixed_points_with_names = [(O1, "O1"), (O2, "O2")]


fig, ax = plt.subplots()
ax.set_aspect("equal")


xdata, ydata = [], []
ln, = plt.plot([], [], 'b-')

vectors = []
points  = []

titles  = []
lines   = []

# Animation initialisation
def init_model():
    ax.set_xlim(-20, 50)
    ax.set_ylim(-15, 45)
    plt.title('Task 1')
    return ln,


# Remove all the previous points, labels, vectors
def clear_screen():
    if len(ax.patches):
        ax.patches[-1].remove()

    while len(points):
        for item in points[-1]:
            item.remove()
        points.pop()

    while len(lines):
        for item in lines[-1]:
            item.remove()
        lines.pop()

    while len(titles):
        titles[-1].remove()
        titles.pop()

    while len(vectors):
        vectors[-1].remove()
        vectors.pop()

# Animation update on each frame
def update_model(state):

    clear_screen()

    # (position, velocity, acceleration) = state
    position = state

    # Draw all moving points of interest on the screen
    points.extend([
        plt.plot(i[0], i[1], marker="o", markersize=3, markeredgecolor="black", markerfacecolor="black")
        for i in position
    ])

    # Draw all fixed points
    points.extend([
        plt.plot(point[0], point[1], marker="o", markersize=3, markeredgecolor="black", markerfacecolor="black")
        for (point, _) in fixed_points_with_names
    ])


    (A, B, S, T, Q, O, M) = position
    # (A_velocity, B_velocity, C_velocity, D_velocity, E_velocity, F_velocity) = velocity
    # (A_acceleration, B_acceleration, C_acceleration, D_acceleration, E_acceleration, F_acceleration) = acceleration


    # Plot all lines on the screen
    lines.extend([
        plt.plot([O1[0], A[0]], [O1[1], A[1]], linestyle="solid", linewidth=2, color="blue"),
        plt.plot([B[0], O2[0]], [B[1], O2[1]], linestyle="solid", linewidth=2, color="blue"),
        plt.plot([S[0], Q[0]], [S[1], Q[1]], linestyle="solid", linewidth=2, color="blue"),
        plt.plot([Q[0], O[0]], [Q[1], O[1]], linestyle="solid", linewidth=2, color="blue"),
        plt.plot([O[0], T[0]], [O[1], T[1]], linestyle="solid", linewidth=2, color="blue"),
        plt.plot([S[0], T[0]], [S[1], T[1]], linestyle="solid", linewidth=2, color="blue"),

    ])

    # Add titles to points on the plot
    titles.extend([
        ax.text(i[0][0], i[0][1] + 3, i[1], fontsize=10)
        for i in zip(position, points_names)
    ])

    # Draw all tiltles of fixed points
    titles.extend([
        ax.text(point[0], point[1] + 3, name, fontsize=10)
        for (point, name) in fixed_points_with_names
    ])

    #     # Draw all the velocity vectors on the plot
    # vectors.extend([
    #     plt.quiver([float(pos[0].evalf())], [float(pos[1].evalf())], [float(vec[0])], [float(vec[1])], color="black", units='xy', scale = 2, scale_units='xy', angles='xy')
    #     for (pos, vec) in zip(position, velocity)
    # ])

    # # Draw all the total acceleration vectors on the plot
    # vectors.extend([
    #     plt.quiver([float(pos[0].evalf())], [float(pos[1].evalf())], [float(acc[0])], [float(acc[1])], color="green", units='xy', scale = 2, scale_units='xy', angles='xy')
    #     for (pos, acc) in zip(position, acceleration)
    # ])

    return ln,


anim = FuncAnimation(fig, update_model, frames=states,
                        init_func=init_model, blit=True)


from IPython.display import HTML, display
display(HTML(anim.to_jshtml()))
plt.close(fig)
```
